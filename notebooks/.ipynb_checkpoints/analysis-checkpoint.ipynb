```python
import geopandas as gpd
import pandas as pd
import rasterio
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
from rasterio.windows import Window
```

```python
import xarray as xr

ds = xr.open_dataset("/Users/luamarty/Desktop/sds/project/data/raw/Landsat/LandsatComposite_Zurich_1985.tif", engine="rasterio")
ds
red = ds.band_data.sel(band=3)
nir = ds.band_data.sel(band=4)
thermal = ds.band_data.sel(band=7)
```

```python
ndvi = (nir - red) / (nir + red)
ndvi = ndvi.where((nir + red) != 0)
ndvi = ndvi.clip(min=0, max=1)

ndvi.name = "NDVI"
ndvi.attrs = {"long_name": "NDVI"}

ndvi.plot(cmap="RdYlGn")
plt.title("NDVI extent")
plt.ylabel("Northing")
plt.xlabel("Easting")
plt.show()
```

```python
thermal_k = thermal         
thermal_c = thermal_k - 273.15

thermal_c.name = "Temperature"
thermal_c.attrs = {"long_name": "Brightness temperature (°C)"}

#ndvi based emissivity
pv = ((ndvi - ndvi.min()) / (ndvi.max() - ndvi.min())) ** 2
emissivity = 0.004 * pv + 0.986

#compute LST
BT = thermal  # already Kelvin

lambda_ = 11.45e-6
rho = 1.438e-2

lst = BT / (1 + (lambda_ * BT / rho) * np.log(emissivity))
lst_c = lst - 273.15

lst_c

lst_c.plot(cmap="plasma")
plt.title("Brightness temperature (°C)")
plt.show()


print("Temp min:", float(lst_c.min()))
print("Temp max:", float(lst_c.max()))
```
